```python
# importing usefull libs
import numpy as np
import pandas as pd
```

```python
# reading datasets
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# displaying how data looks like by only 1 value
movies.head(4)
```

```python
credits.head(4)
```

```python
# movies.merge(credits, on='title').shape to check the number of items and columns
movie_dataset = movies.merge(credits, on='title')
```

```python
movie_dataset
```

```python
# genres, id, keywords, title, overview, cast, crew
movie_dataset = movie_dataset[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movie_dataset.info()
```

```python
movie_dataset.head(4)
```

```python
# checking if there is any missing values 
movie_dataset.isnull().sum()
```

```python
# There was 3 missing values in the overview so i removed it
movie_dataset.dropna(inplace=True)
```

```python
# checking if there is any duplicate values
movie_dataset.duplicated().sum()
```

```python
movie_dataset.iloc[0].genres
```

```python
# converting genres to a list because it in a werid value
import ast

def conv(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movie_dataset['genres'] = movie_dataset['genres'].apply(conv)
```

```python
movie_dataset.head(1)
# done with genres, conv into a list
```

```python
# now we do this thing with other columns
movie_dataset['keywords'] = movie_dataset['keywords'].apply(conv)
```

```python
movie_dataset.head(1)
```

```python
movie_dataset['cast'][0]
```

```python
def convCast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count+1
        else:
            break
    return L
```

```python
movie_dataset['cast'] = movie_dataset['cast'].apply(convCast)
```

```python
movie_dataset.head(1)
```

```python
# crew is difficult to clean or convert into the list
def getDirect(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movie_dataset['crew'] = movie_dataset['crew'].apply(getDirect)
```

```python
# now converting overview string into list and split every words
movie_dataset['overview'] = movie_dataset['overview'].apply(lambda x:x.split())
```

```python
movie_dataset.head(1)
```

```python
# now we will remove the spaces from the name of cast and crew because there can problems between "Zoe saldana" and "Zoe jaha" 
# same things we will apply in the genres "science fiction"
movie_dataset['genres'] = movie_dataset['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movie_dataset['keywords'] = movie_dataset['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movie_dataset['crew'] = movie_dataset['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movie_dataset['cast'] = movie_dataset['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movie_dataset.head(1)
```

```python
# now we make a single column and merge overview, genres, keywords, cast, crew
movie_dataset['tags'] = movie_dataset['overview'] + movie_dataset['genres'] + movie_dataset['cast'] + movie_dataset['crew']
```

```python
movie_dataset.head(4)
```

```python
new_df = movie_dataset[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(10)
```

```python
new_df['tags'][0]
```

```python
# now convert into lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(3)
```

```python
import nltk
```

```python
new_df['tags'][2000]
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
def stem(text):
    new_list = []
    for item in text.split():
        new_list.append(porter_stemmer.stem(item))
    return " ".join(new_list)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# now we have to find similarity scoure
we have to find how two movies are similer to each and give a scoure to it

-> convert the tags text into vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
# stemming to remove similer words loving, loved => love

# did it above using stem()
```

```python
# now we will calculate cosine distance between each movie
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend_movie_function(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    dis = similarity[movie_index]
    movie_list = sorted(list(enumerate(dis)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend_movie_function('The Avengers')
```
